Medidas de Dispersión

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cars.csv')
df
```

```
# Calculamos la desviación estándar
df['price_usd'].std()
```

```
# Obtenemos el Rango = valor max - valor min
rango = df['price_usd'].max() -df['price_usd'].min()
rango
```

```
# Calculamos Quartiles  Cuartiles

median = df['price_usd'].median() # Uno de los cuartiles es la mediana
Q1 = df['price_usd'].quantile(q=0.25) # Primer quartile, es el 25%, un 1/4 de los datos
Q3 = df['price_usd'].quantile(q=0.75) # Tercer quartile, es el 75% de los datos, las 3/4 partes
min_val = df['price_usd'].quantile(q=0)
max_val = df['price_usd'].quantile(q=1.0)
print(min_val, Q1, median, Q3, max_val)
```

```
iqr = Q3 - Q1 # Calculamos el iqr rango intercuartil
iqr
```

### Límites para detección de outliers (datos simetricamente distribuidos)

Datos entre $$Q_1 -1.5 \times \text{IQR}$$ y $$Q_3 + 1.5 \times \text{IQR}$$

```
minlimit = Q1 - 1.5*iqr
maxlimit = Q3 + 1.5*iqr
print(minlimit, maxlimit)
```
