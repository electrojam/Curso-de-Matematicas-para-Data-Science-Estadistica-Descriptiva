```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

iris.head()
```

```
# Graficamos scatterplot

sns.scatterplot(data=iris, x = 'sepal_length', y = 'petal_length')
```

```
# Graficamos scatterplot discriminando por especies

sns.scatterplot(data=iris, x = 'sepal_length', y = 'petal_length', hue = 'species')
```

```
# Graficamos dispersin con joinplot de seaborn

sns.jointplot(data=iris, x = 'sepal_length', y = 'petal_length')
```

```
# Graficamos  dispersin con joinplot de seaborn discriminando por especies

sns.jointplot(data=iris, x = 'sepal_length', y = 'petal_length', hue = 'species')
```

```
# box plot

sns.boxplot(data = iris, x = 'species', y = 'sepal_length' )
```

```
# reto scarlet plot

sns.set_theme()
sns.jointplot(data=iris, x = 'sepal_width', y = 'petal_width', hue = 'species', palette='inferno') 

```

```
# Bar plot

sns.set_theme()
sns.barplot(data=iris, x = 'species', y = 'petal_length', palette='inferno') 

```

```
iris.groupby('species').mean()
```

```
sns.barplot(data=iris, x='species', y='petal_length')
```

```
species_mean_df = iris.groupby('species').mean().reset_index()
species_mean_df
```

```
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (7,7))

sns.barplot(ax = axes[0,0], data = species_mean_df, x = 'species', y = 'petal_length')
sns.barplot(ax = axes[0,1], data = species_mean_df, x = 'species', y = 'sepal_length')
sns.barplot(ax = axes[1,0], data = species_mean_df, x = 'species', y = 'petal_width')
sns.barplot(ax = axes[1,1], data = species_mean_df, x = 'species', y = 'sepal_length')
```
